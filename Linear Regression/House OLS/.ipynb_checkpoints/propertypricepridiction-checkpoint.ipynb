```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('HousePrices.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(["PoolQC","Fence","MiscFeature","Alley"],axis=1,inplace=True)
```

```python
X = df.drop("Property_Sale_Price", axis=1)
y = df["Property_Sale_Price"]
```

```python
fig, ax = plt.subplots(figsize=(30,30))
sns.heatmap(df.corr(),annot=True)
```

```python
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,X.columns)])

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
X_train, X_test , y_train, y_test = train_test_split(transformed_X,
                                                     y,
                                                     test_size=0.2,
                                                     random_state=0)
```

```python
reg = RandomForestRegressor()
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train),reg.score(X_test,y_test)
```

```python
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
linreg.score(X_train,y_train),linreg.score(X_test,y_test)
```
